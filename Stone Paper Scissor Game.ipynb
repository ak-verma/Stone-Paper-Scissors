```python
from random import randint
i=1;
d=e=0;
while (i<2):
    i=i+1;
    a=input('rock(r) or paper(p) or scissors(s)?\n')
    print('You chose' + '\t', a)
    b=randint(1,3)
    if b==1:
        b='r'
    elif b==2:
        b='p'
    else: 
        b='s'
    print("My choice",b,'\n')
    
    if ((a=='r' and b=='p') or(a=='s' and b=='r')or(a=='p'and b=='s')):
        print('You win\n')
        d=d+1;
    elif a==b:
        print('Draw\n')
    else:
        print('You lose\n')
        e=e+1;
    
    print('Another battle?\n')
    c=input('yes(y) OR no(n)\n')
    
    if c=='y':
        i=1;
    else:
        print('See you again')
    

print('Your score\n',d)
print('My score\n',e)
if d>e:
    print('You win the tournament')
elif d==e:
    print('Oops it\'s a draw')
else:
    print('I am the champion')
```
